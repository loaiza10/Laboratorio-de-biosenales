1. Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo n1≤n0≤n2

* Escalón unitario
    El escalón unitario se define como:

    u(n − n0) = {1 si n ≥ n0 - 0 si n < n0

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
def escalon_unitario(n_1,n_2, n_0):
    """Genera un escalón unitario u(n - n0)
    definido en el intervalo [n1, n2]"""

    #indices 
    n = np.arange(n_1, n_2 + 1)
    u = np.where(n >= n_0, 1, 0)
    return n, u

n, u = escalon_unitario(-10, 10, 2)
plt.stem(n, u)
plt.title("Escalón unitario u(n - n0)")
plt.xlabel("n")
plt.ylabel("u(n)")
plt.grid()
plt.show()
```

2. Implemente una función en Python que permita generar una función rampa, definida en un intervalo 𝑛1≤𝑛0≤𝑛2 con una pendiente m.

Rampa

La función rampa se define como:

𝑟(𝑛 − 𝑛0) = {
0 𝑠𝑖 𝑛 < 𝑛0

                𝑛 − 𝑛0 𝑠𝑖 𝑛 ≥ 𝑛0


```python
n = np.array([-2,-1,0,1,2,3])
n0 = 1

x = n >= n0
print(x) 
print(x.astype(int))

```

```python
def rampa(n0,n1,n2,m): # n0: punto de corte,  n1: punto inicial, n2: punto final, m: pendiente

    n = np.arange(n1,n2+1) # vector de n1 hasta n2 + 1 para que incluya el ultimo numero del rango.
    x = (n-n0)*m*(n>=n0) # multplicamos el vector boleano y el vector original

    return x,n


x,n = rampa(0,-5,5,2)

plt.stem(n,x)
plt.plot(n,x)
plt.show()
```

3. Implemente una función en Python que permita generar una función exponencial compleja, definida en un intervalo n1≤n0≤n2.
* Exponenciales complejas

    Las exponenciales complejas son de la forma:
    x(n) = e
    (σ+jω0)n, ∀n

    En la que σ produce una atenuación si es negativo o una amplificación si es positivo. En Python
se hace uso del comando np.exp.

```python
def exponencial_compleja(n1, n2, sigma, w0):
    """
    Genera la señal exponencial compleja:
    x(n) = exp((sigma + j*w0)*n)
    
    Parámetros:
        n1 : inicio del intervalo
        n2 : fin del intervalo
        sigma : parte real (crecimiento o atenuación)
        w0 : frecuencia angular
        
    Retorna:
        n : vector de índices
        x : señal exponencial compleja
    """
    
    # Vector discreto
    n = np.arange(n1, n2 + 1)
    
    # Exponencial compleja
    x = np.exp((sigma + 1j*w0) * n)
    
    return n, x

n, x = exponencial_compleja(-10, 10, sigma=-0.1, w0=0.5)

plt.stem(n, np.real(x))
plt.title("Parte Real de la Exponencial Compleja")
plt.xlabel("n")
plt.ylabel("Re{x(n)}")
plt.grid()
plt.show()

plt.figure()
plt.stem(n, np.abs(x))
plt.title("Magnitud |x(n)|")
plt.grid()
plt.show()

plt.figure()
plt.stem(n, np.imag(x))
plt.title("Parte Imaginaria")
plt.grid()
plt.show()
```

**4. Implemente una función en Python que permita generar una función sinusoidal, definida en un intervalo 𝑛1≤𝑛0≤𝑛2.**

Sinusoides

Estas secuencias son de la forma:

*𝑥(𝑛) = sin (𝜔0𝑛 + 𝜃0)*

```python
def seno(w0,theta,n1,n2):

    n = np.arange(n1,n2+1)
    
    x = np.sin(w0*n + theta)
    
    return x,n
```

```python
x,n = seno(0.2*np.pi, 0, -20, 50)
plt.stem(n,x)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Señal sinusoidal discreta")
plt.grid(True)
plt.show()
```
