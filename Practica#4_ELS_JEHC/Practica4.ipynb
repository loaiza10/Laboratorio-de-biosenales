### **Esquema del registro de biopotenciales**

El proceso de adquisición de biopotenciales comprende una serie de etapas que van desde la generación de los potenciales de acción en las células excitables hasta su digitalización y análisis por un sistema electrónico [1]. Estas etapas pueden representarse como: 


El esquema inicia en el sistema biológico, donde las células excitables generan biopotenciales como resultado de procesos electroquímicos. Estos potenciales se propagan a través del tejido hasta ser detectados por electrodos colocados en la superficie del cuerpo.

Posteriormente, la señal captada pasa por una etapa de acondicionamiento, donde se amplifica y filtra para mejorar su calidad y reducir el ruido. Luego, la señal analógica es convertida a formato digital mediante un convertidor A/D, permitiendo su procesamiento computacional.

Finalmente, la señal digital es analizada, visualizada y almacenada, facilitando su interpretación clínica o experimental [1], [2].

### **Etapas del proceso**

* **Generación del biopotencial**
Los biopotenciales se originan debido a la actividad electroquímica de células excitables, como neuronas y células musculares. Estos potenciales son consecuencia del movimiento de iones a través de la membrana celular, generando potenciales de acción que constituyen la base de señales como el ECG y EMG [1].

* **Propagación en el tejido biológico**
Una vez generado, el potencial de acción se propaga a través del tejido biológico. En el caso del corazón, esta propagación ocurre de manera organizada para producir el ciclo cardíaco, mientras que en el músculo esquelético se asocia a la contracción muscular [1].

* **Captación mediante electrodos**
Los electrodos colocados sobre la piel permiten medir diferencias de potencial eléctrico generadas por la actividad fisiológica. Estos pueden configurarse en derivaciones bipolares o monopolares, dependiendo de la aplicación. La correcta colocación y preparación de la piel es fundamental para obtener señales de calidad [1].

* **Acondicionamiento de la señal**
Las señales biológicas presentan baja amplitud (del orden de microvoltios a milivoltios) y son susceptibles al ruido. Por ello, se requiere una etapa de acondicionamiento que incluye amplificación, filtrado y aislamiento eléctrico para mejorar la relación señal-ruido [2].

* **Conversión analógica-digital (ADC)**
La señal analógica acondicionada debe convertirse en formato digital mediante un convertidor A/D. Este proceso implica muestreo y cuantificación. En sistemas como el utilizado en la práctica, el ADC se encuentra integrado en el microcontrolador (por ejemplo, Arduino), permitiendo la posterior manipulación digital de la señal [1],[2].

* **Procesamiento y visualización**
Finalmente, la señal digital es procesada mediante software para su análisis e interpretación. Esto puede incluir filtrado digital, detección de eventos relevantes (como picos R en ECG) y visualización gráfica de la señal [2].

### Componentes de un sistema de adquisición de biopotenciales

Un sistema típico de adquisición de biopotenciales está compuesto por:

* **Electrodos (sensores):** encargados de captar la señal biológica [1].

* **Etapa de acondicionamiento:** incluye amplificadores de instrumentación y filtros analógicos [2].

* **Convertidor analógico-digital (ADC):** digitaliza la señal para su procesamiento [1],[2].

* **Unidad de procesamiento:** puede ser un microcontrolador o un computador [1].

* **Sistema de visualización y almacenamiento:** permite analizar y guardar la información [2].

* **Sistema de comunicación:** transmite los datos (por ejemplo, comunicación serial) [2].

### **3. Describa ¿Cuál es la problemática asociada al acoplamiento piel-electrodo? ¿Cuáles estrategias se utilizan para mejorarla? Además, responda ¿Por qué el acoplamiento piel-electrodo constituye la etapa más crítica en el registro de señales de superficie (sEMG, EEG, ECG)?**

### **Acoplamiento piel–electrodo**

El acoplamiento piel–electrodo es una de las etapas más críticas en la adquisición de biopotenciales, ya que constituye la interfaz entre el sistema biológico y el sistema electrónico de medición. La principal problemática asociada a este acoplamiento radica en que la piel no es un conductor ideal, debido a la presencia de la capa córnea, la cual presenta alta resistencia eléctrica y comportamiento no lineal. Esto genera una impedancia de contacto elevada e inestable, afectando la calidad de la señal adquirida.

Además, el acoplamiento puede introducir diversos problemas, tales como:

* **Ruido e interferencias**, especialmente por movimiento (artefactos de movimiento).
* **Variaciones en la impedancia**, causadas por sudoración, temperatura o presión del electrodo.
* **Polarización del electrodo**, que genera voltajes espurios en la interfaz.
* **Baja relación señal-ruido**, debido a la atenuación de señales de baja amplitud como ECG o EMG.

### **Estrategias para mejorar el acoplamiento piel–electrodo**

Para minimizar estos efectos y mejorar la calidad de la señal, se emplean diversas estrategias:

* **Preparación de la piel**: limpieza con alcohol para eliminar grasa, sudor y células muertas, reduciendo la impedancia [1].

* **Uso de gel conductor**: mejora la conductividad eléctrica entre la piel y el electrodo, disminuyendo la resistencia de contacto.

* **Electrodos adecuados (Ag/AgCl)**: reducen la polarización y proporcionan una interfaz más estable.

* **Fijación adecuada de los electrodos**: evita movimientos relativos entre piel y electrodo, reduciendo artefactos.

* **Ubicación correcta**: colocar los electrodos en zonas con buena conductividad y menor interferencia muscular (según el tipo de señal).

* **Uso de amplificadores de alta impedancia de entrada**: disminuyen el efecto de la impedancia de contacto sobre la medición [2].

### **¿Por qué es la etapa más crítica?**

El acoplamiento piel–electrodo se considera la etapa más crítica en el registro de biopotenciales porque es el primer punto de contacto entre la señal fisiológica y el sistema de medición. Cualquier error, ruido o distorsión introducida en esta etapa no puede ser completamente eliminada en etapas posteriores, ni siquiera mediante procesamiento digital.

Dado que las bioseñales tienen amplitudes muy pequeñas (microvoltios a milivoltios), una mala interfaz piel–electrodo puede degradar significativamente la señal, comprometiendo la precisión del análisis y la validez de los resultados. Por esta razón, garantizar un buen acoplamiento es fundamental para obtener registros confiables en técnicas como ECG y EMG [1], [2].

### **Aplicaciones del ECG (Electrocardiografía)**

* **Diagnóstico de arritmias cardíacas**
El ECG permite identificar alteraciones en el ritmo cardíaco, como fibrilación auricular o taquicardias.

* **Tipo de registro:** ECG en reposo o monitoreo continuo (Holter).
* **Procesamiento:** filtrado de ruido (50/60 Hz), detección de picos R, análisis de intervalos RR y morfología de ondas (P, QRS, T) para identificar irregularidades [1], [2].

* **Detección de isquemia e infarto de miocardio**
Se utiliza para identificar cambios en el segmento ST y en la onda T, indicadores de daño en el tejido cardíaco.

* **Tipo de registro:** ECG de 12 derivaciones.
* **Procesamiento:** filtrado, segmentación de latidos, análisis del segmento ST y comparación con patrones normales para detectar elevaciones o depresiones características [1], [2].

* **Análisis de la variabilidad de la frecuencia cardíaca (HRV)**
Permite evaluar la regulación del sistema nervioso autónomo y su relación con condiciones como estrés o enfermedades cardiovasculares.

* **Tipo de registro:** ECG continuo (reposo o monitoreo prolongado).
* **Procesamiento:** detección de picos R, cálculo de intervalos RR y análisis estadístico en el dominio del tiempo y frecuencia (ej. SDNN, RMSSD) [2].

### **Aplicaciones del EMG (Electromiografía)**

* **Diagnóstico de trastornos neuromusculares**
El EMG permite evaluar enfermedades como neuropatías, miopatías o esclerosis lateral amiotrófica (ELA).

* **Tipo de registro:** EMG invasivo (aguja) o de superficie (sEMG).
* **Procesamiento:** filtrado, rectificación de la señal y análisis de amplitud y frecuencia para identificar patrones anormales de activación muscular [1], [2].

* **Evaluación de la función muscular y rehabilitación**
Se emplea para analizar la activación muscular en pacientes en procesos de recuperación (por ejemplo, después de una lesión).

* **Tipo de registro:** EMG de superficie (sEMG).
* **Procesamiento:** filtrado, cálculo de la envolvente (RMS), análisis temporal de la activación muscular y comparación entre músculos o sesiones [2].

* **Control de prótesis y sistemas de asistencia**
Las señales EMG se utilizan como señales de control para prótesis mioeléctricas o dispositivos de asistencia.

* **Tipo de registro:** EMG de superficie multicanal.
* **Procesamiento:** filtrado, extracción de características (RMS, media absoluta, frecuencia media) y clasificación mediante algoritmos (machine learning) para traducir la señal en comandos de movimiento [2], [3].
