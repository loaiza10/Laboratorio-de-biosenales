### **1. Esquema del registro de biopotenciales**

El proceso de adquisición de biopotenciales comprende una serie de etapas que van desde la generación de los potenciales de acción en las células excitables hasta su digitalización y análisis por un sistema electrónico. Estas etapas pueden representarse como: 


### **Etapas del proceso**

* **Generación del biopotencial**
Los biopotenciales se originan debido a la actividad electroquímica de células excitables, como neuronas y células musculares. Estos potenciales son consecuencia del movimiento de iones a través de la membrana celular, generando potenciales de acción que constituyen la base de señales como el ECG y EMG

* **Propagación en el tejido biológico**
Una vez generado, el potencial de acción se propaga a través del tejido biológico. En el caso del corazón, esta propagación ocurre de manera organizada para producir el ciclo cardíaco, mientras que en el músculo esquelético se asocia a la contracción muscular

* **Captación mediante electrodos**
Los electrodos colocados sobre la piel permiten medir diferencias de potencial eléctrico generadas por la actividad fisiológica. Estos pueden configurarse en derivaciones bipolares o monopolares, dependiendo de la aplicación. La correcta colocación y preparación de la piel es fundamental para obtener señales de calidad.

* **Acondicionamiento de la señal**
Las señales biológicas presentan baja amplitud (del orden de microvoltios a milivoltios) y son susceptibles al ruido. Por ello, se requiere una etapa de acondicionamiento que incluye amplificación, filtrado y aislamiento eléctrico para mejorar la relación señal-ruido [3].

* **Conversión analógica-digital (ADC)**
La señal analógica acondicionada debe convertirse en formato digital mediante un convertidor A/D. Este proceso implica muestreo y cuantificación. En sistemas como el utilizado en la práctica, el ADC se encuentra integrado en el microcontrolador (por ejemplo, Arduino), permitiendo la posterior manipulación digital de la señal

* **Procesamiento y visualización**
Finalmente, la señal digital es procesada mediante software para su análisis e interpretación. Esto puede incluir filtrado digital, detección de eventos relevantes (como picos R en ECG) y visualización gráfica de la señal [3].

### Componentes de un sistema de adquisición de biopotenciales

Un sistema típico de adquisición de biopotenciales está compuesto por:

* **Electrodos (sensores):** encargados de captar la señal biológica.

* **Etapa de acondicionamiento:** incluye amplificadores de instrumentación y filtros analógicos.

* **Convertidor analógico-digital (ADC):** digitaliza la señal para su procesamiento.

* **Unidad de procesamiento:** puede ser un microcontrolador o un computador.

* **Sistema de visualización y almacenamiento:** permite analizar y guardar la información.

* **Sistema de comunicación:** transmite los datos (por ejemplo, comunicación serial) [3].
