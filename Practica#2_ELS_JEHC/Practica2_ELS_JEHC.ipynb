1. Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo n1≤n0≤n2

* Escalón unitario
    El escalón unitario se define como:

    u(n − n0) = {1 si n ≥ n0 - 0 si n < n0

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
def escalon_unitario(n_1,n_2, n_0):
    """Genera un escalón unitario u(n - n0)
    definido en el intervalo [n1, n2]"""

    #indices 
    n = np.arange(n_1, n_2 + 1)
    u = np.where(n >= n_0, 1, 0)
    return n, u

n, u = escalon_unitario(-10, 10, 2)
plt.stem(n, u)
plt.title("Escalón unitario u(n - n0)")
plt.xlabel("n")
plt.ylabel("u(n)")
plt.grid()
plt.show()
```

2. Implemente una función en Python que permita generar una función rampa, definida en un intervalo 𝑛1≤𝑛0≤𝑛2 con una pendiente m.

Rampa

La función rampa se define como:

𝑟(𝑛 − 𝑛0) = {
0 𝑠𝑖 𝑛 < 𝑛0

                𝑛 − 𝑛0 𝑠𝑖 𝑛 ≥ 𝑛0


```python
n = np.array([-2,-1,0,1,2,3])
n0 = 1

x = n >= n0
print(x) 
print(x.astype(int))

```

```python
def rampa(n0,n1,n2,m): # n0: punto de corte,  n1: punto inicial, n2: punto final, m: pendiente

    n = np.arange(n1,n2+1) # vector de n1 hasta n2 + 1 para que incluya el ultimo numero del rango.
    x = (n-n0)*m*(n>=n0) # multplicamos el vector boleano y el vector original

    return x,n


x,n = rampa(0,-5,5,2)

plt.stem(n,x)
plt.plot(n,x)
plt.show()
```

3. Implemente una función en Python que permita generar una función exponencial compleja, definida en un intervalo n1≤n0≤n2.
* Exponenciales complejas

    Las exponenciales complejas son de la forma:
    x(n) = e
    (σ+jω0)n, ∀n

    En la que σ produce una atenuación si es negativo o una amplificación si es positivo. En Python
se hace uso del comando np.exp.

```python
def exponencial_compleja(n1, n2, sigma, w0):
    """
    Genera la señal exponencial compleja:
    x(n) = exp((sigma + j*w0)*n)
    
    Parámetros:
        n1 : inicio del intervalo
        n2 : fin del intervalo
        sigma : parte real (crecimiento o atenuación)
        w0 : frecuencia angular
        
    Retorna:
        n : vector de índices
        x : señal exponencial compleja
    """
    
    # Vector discreto
    n = np.arange(n1, n2 + 1)
    
    # Exponencial compleja
    x = np.exp((sigma + 1j*w0) * n)
    
    return n, x

n, x = exponencial_compleja(-10, 10, sigma=-0.1, w0=0.5)

plt.stem(n, np.real(x))
plt.title("Parte Real de la Exponencial Compleja")
plt.xlabel("n")
plt.ylabel("Re{x(n)}")
plt.grid()
plt.show()

plt.figure()
plt.stem(n, np.abs(x))
plt.title("Magnitud |x(n)|")
plt.grid()
plt.show()

plt.figure()
plt.stem(n, np.imag(x))
plt.title("Parte Imaginaria")
plt.grid()
plt.show()
```

```python
# Intervalo pedido
n = np.arange(-10, 11)

# Escalón
def u(n, n0):
    return np.where(n >= n0, 1, 0)

# Rampa
def r(n, n0):
    return (n - n0) * u(n, n0)

# Señal
x4 = 2*r(n, -3) - r(n, 2) - 5*u(n, 3)

# Gráfica
plt.stem(n, x4)
plt.title("x4(n)")
plt.grid()
plt.show()
```

La señal es por tramos.
Tiene comportamientos distintos dependiendo del intervalo de n.

Las señales discretas pueden construirse usando funciones básicas:

Impulso δ(n)

Escalón u(n)

Rampa r(n)

* Estas funciones permiten modelar comportamientos complejos mediante combinaciones simples.
* El uso de NumPy facilita implementar estas señales mediante operaciones vectorizadas.
* La gráfica stem es la más adecuada para representar señales discretas en tiempo.

* La señal es no periódica.
* Tiene tres comportamientos diferentes:

Primer tramo
−5≤n≤0
Crecimiento lineal hasta el valor máximo 5.

Segundo tramo
0≤n≤5
Decrece hasta 0.

Tercer tramo
5≤n≤10
Vuelve a crecer.

Cuarto tramo
10≤n≤14

Quinto tramo
15≤n≤18

```python
# Nuevo intervalo más grande para evitar pérdidas
n_ext = np.arange(-20, 25)

# Inicializamos en cero
x_ext = np.zeros_like(n_ext)

# Ubicar x(n) dentro del nuevo intervalo
for i in range(len(n)):
    idx = np.where(n_ext == n[i])[0][0]
    x_ext[idx] = x[i]

# Desplazamiento x(n-4)
x_shift = np.zeros_like(n_ext)
for i in range(len(n_ext)):
    if (n_ext[i]-4) in n_ext:
        idx = np.where(n_ext == n_ext[i]-4)[0][0]
        x_shift[i] = x_ext[idx]

# x5
x5 = 2*x_shift + x_ext

plt.stem(n_ext, x5)
plt.title("x5(n) = 2x(n-4) + x(n)")
plt.grid()
plt.show()
```

* La señal es más grande en amplitud, porque incluye el término 2x(n-4).
* Aparecen valores altos cuando:

la señal original y la señal desplazada coinciden en el mismo índice.

En el tramo final aparecen valores cercanos a 20, porque:
10+2(10)=30
pero dependiendo de los solapamientos se observan valores cercanos a 20.

* El desplazamiento genera una superposición de señales.
* Este tipo de operación aparece mucho en:

convolución
