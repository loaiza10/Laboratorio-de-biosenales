### **Introducción**

El análisis frecuencial constituye una de las herramientas fundamentales en el procesamiento digital de señales biomédicas, ya que permite estudiar la distribución energética de una señal en función de la frecuencia y comprender fenómenos fisiológicos o patológicos asociados a determinadas bandas espectrales. Técnicas como la Transformada Discreta de Fourier (DFT), la Transformada Inversa de Fourier (IDFT) y el periodograma de Welch son ampliamente utilizadas en bioingeniería para el análisis de señales como EEG y ECG [1].

La DFT permite transformar una señal desde el dominio temporal hacia el dominio frecuencial, facilitando la identificación de las componentes sinusoidales presentes en la señal. A su vez, la IDFT posibilita reconstruir la señal original a partir de su espectro frecuencial, demostrando la equivalencia entre ambas representaciones [2].

En bioseñales, el análisis espectral permite identificar ruido, artefactos y componentes fisiológicamente relevantes. En señales ECG, por ejemplo, gran parte de la información útil se encuentra entre 0.5 y 40 Hz, mientras que frecuencias inferiores suelen asociarse a deriva de línea base y frecuencias superiores a interferencia electromagnética o ruido muscular [3]. De igual manera, en señales EEG el análisis frecuencial permite identificar alteraciones neuronales relacionadas con enfermedades neurodegenerativas como el Parkinson [4].

En esta práctica se implementaron diferentes técnicas de análisis frecuencial aplicadas inicialmente a señales sinusoidales sintéticas y posteriormente a señales biomédicas reales de ECG y EEG. Se realizó el cálculo de la DFT e IDFT, extracción espectral de componentes sinusoidales, filtrado en el dominio de la frecuencia y estimación de densidad espectral de potencia mediante el método de Welch, con el fin de comprender el comportamiento espectral de diferentes bioseñales.

### **Análisis y Discusión de Resultados**

**Señal compuesta y criterio de Nyquist**

Inicialmente se construyó una señal compuesta por tres componentes sinusoidales de 40, 80 y 160 Hz. De acuerdo con el teorema de muestreo de Nyquist-Shannon, la frecuencia mínima de muestreo necesaria para evitar aliasing corresponde al doble de la frecuencia máxima presente en la señal:

$$F_s \geq 2f_{max}$$

Dado que la frecuencia máxima fue 160 Hz, la frecuencia mínima teórica fue 320 Hz. Sin embargo, se utilizó una frecuencia de muestreo de 1600 Hz para obtener una representación temporal más adecuada y mejorar la resolución espectral.

La Transformada Discreta de Fourier permitió identificar correctamente las tres componentes sinusoidales presentes en la señal. Los picos espectrales obtenidos en aproximadamente 40, 80 y 160 Hz confirmaron la correcta implementación de la DFT y la correspondencia entre el dominio temporal y frecuencial.

Posteriormente, la aplicación de la Transformada Inversa de Fourier permitió reconstruir prácticamente de manera exacta la señal original. El error cuadrático medio obtenido fue cercano a cero, confirmando que la DFT y la IDFT constituyen operaciones inversas entre sí.

**Filtrado en el dominio de la frecuencia**

El filtrado espectral realizado mediante anulación selectiva de coeficientes de la DFT demostró que es posible implementar filtros ideales directamente en el dominio frecuencial. Al conservar únicamente determinados coeficientes espectrales y anular el resto, se logró extraer cada una de las componentes sinusoidales individuales de la señal compuesta.

Este procedimiento equivale a aplicar un filtro pasa-banda ideal cuya respuesta frecuencial corresponde a una ventana rectangular. La reconstrucción mediante IDFT permitió verificar que las componentes extraídas correspondían efectivamente a las sinusoides originales de 40, 80 y 160 Hz.

**Análisis frecuencial de la señal ECG**

El análisis espectral de la señal ECG mediante el periodograma de Welch permitió identificar la distribución energética de la señal y detectar componentes de ruido presentes en el registro.

Para la estimación espectral se utilizó una ventana de Hanning con un solapamiento del 50%. La ventana Hanning reduce la fuga espectral producida por discontinuidades en los extremos de cada segmento, mientras que el solapamiento mejora la estabilidad estadística del estimador espectral [5].

La resolución frecuencial utilizada estuvo dada por:

$$\Delta f = \frac{F_s}{N}$$

donde $F_s$ corresponde a la frecuencia de muestreo y $N$ al número de muestras por segmento.

El periodograma mostró que la mayor parte de la energía útil del ECG se encontraba entre 0.5 y 40 Hz. Además, se identificaron componentes de baja frecuencia asociadas a deriva de línea base y posibles componentes de alta frecuencia relacionadas con ruido electromagnético o muscular.

A partir de ello se implementó un filtro pasa-banda ideal entre 0.5 y 40 Hz en el dominio de la frecuencia. El segundo periodograma de Welch confirmó la eliminación satisfactoria de las componentes de ruido, observándose una reducción significativa de potencia fuera de la banda de interés fisiológico.

**Análisis frecuencial de señales EEG**

Finalmente, se realizó el análisis espectral de señales EEG correspondientes a un sujeto sano y un paciente con enfermedad de Parkinson utilizando el método de Welch sobre los canales motores C3, C4 y Cz.

Los periodogramas obtenidos mostraron diferencias importantes entre ambos sujetos. El paciente con Parkinson presentó un incremento considerable de potencia en bajas frecuencias, especialmente por debajo de 10 Hz, mientras que el sujeto sano presentó una distribución espectral más atenuada y estable.

Este comportamiento es consistente con fenómenos de lentificación cortical reportados en enfermedades neurodegenerativas, donde aumenta la actividad en bandas lentas como delta y theta [6]. Las diferencias fueron particularmente evidentes en los canales motores C3 y C4, lo cual resulta coherente con la afectación motora característica de la enfermedad de Parkinson.

Además, el canal Cz mostró alteraciones similares, sugiriendo una posible afectación funcional más generalizada en regiones corticales motoras. Los resultados obtenidos demuestran que el análisis espectral mediante EEG constituye una herramienta útil para la identificación de alteraciones neuronales asociadas a trastornos neurodegenerativos.

### **Conclusiones**

- La Transformada Discreta de Fourier permitió identificar correctamente las componentes frecuenciales presentes en señales sinusoidales compuestas.
- La Transformada Inversa de Fourier reconstruyó adecuadamente la señal original, verificando la equivalencia entre las representaciones temporal y frecuencial.
- El filtrado en el dominio de la frecuencia permitió extraer selectivamente componentes sinusoidales específicas mediante la anulación de coeficientes espectrales.
- El periodograma de Welch proporcionó estimaciones estables y robustas de la densidad espectral de potencia en señales biomédicas.
- La ventana de Hanning redujo la fuga espectral y el solapamiento del 50% permitió disminuir la varianza del estimador espectral.
- En la señal ECG se identificaron componentes de ruido fuera de la banda fisiológica de interés, las cuales fueron eliminadas exitosamente mediante filtrado espectral.
- Las señales EEG del paciente con Parkinson mostraron incremento de potencia en bajas frecuencias respecto al sujeto sano, evidenciando alteraciones espectrales compatibles con procesos neurodegenerativos.
- El análisis frecuencial constituye una herramienta fundamental en el procesamiento y caracterización de bioseñales.
